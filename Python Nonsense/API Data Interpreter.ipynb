```python
banList = []
banNumbers = []

for ban in df['Ban1']:
    if ban in banList:
        banNumbers[banList.index(ban)] += 1
    else:
        banList.append(ban)
        banNumbers.append(1)
        
for ban in df['Ban2']:
    if ban in banList:
        banNumbers[banList.index(ban)] += 1
    else:
        banList.append(ban)
        banNumbers.append(1)
        
for ban in df['Ban3']:
    if ban in banList:
        banNumbers[banList.index(ban)] += 1
    else:
        banList.append(ban)
        banNumbers.append(1)
        
for ban in df['Ban4']:
    if ban in banList:
        banNumbers[banList.index(ban)] += 1
    else:
        banList.append(ban)
        banNumbers.append(1)
        
for ban in df['Ban5']:
    if ban in banList:
        banNumbers[banList.index(ban)] += 1
    else:
        banList.append(ban)
        banNumbers.append(1)
        
for ban in df['Ban6']:
    if ban in banList:
        banNumbers[banList.index(ban)] += 1
    else:
        banList.append(ban)
        banNumbers.append(1)
        
for ban in df['Ban7']:
    if ban in banList:
        banNumbers[banList.index(ban)] += 1
    else:
        banList.append(ban)
        banNumbers.append(1)
        
for ban in df['Ban8']:
    if ban in banList:
        banNumbers[banList.index(ban)] += 1
    else:
        banList.append(ban)
        banNumbers.append(1)

num1Index = 0
while num1Index < len(banNumbers) - 1:
    num2Index = num1Index + 1
    while num2Index < len(banNumbers):
        num1 = banNumbers[num1Index]
        num2 = banNumbers[num2Index]
        if num1 < num2:
            banNumbers[num1Index], banNumbers[num2Index] = banNumbers[num2Index], banNumbers[num1Index]
            banList[num1Index], banList[num2Index] = banList[num2Index], banList[num1Index]
        num2Index += 1
    num1Index += 1

for ban in banList:
    print(ban)
    print(banNumbers[banList.index(ban)])
print(len(banList))
```

```python
pickList = []
pickNumbers = []
winNumbers = []
iterator = 0


for pick in df['God name']:
    if pick in pickList:
        pickNumbers[pickList.index(pick)] += 1
        if df['Win Status'].values[iterator] == '"Winner"':
            winNumbers[pickList.index(pick)] += 1
    else:
        pickList.append(pick)
        pickNumbers.append(1)
        if df['Win Status'].values[iterator] == '"Winner"':
            winNumbers.append(1)
        else:
            winNumbers.append(0)
    iterator += 1
        
num1Index = 0
while num1Index < len(pickNumbers) - 1:
    num2Index = num1Index + 1
    while num2Index < len(pickNumbers):
        num1 = pickNumbers[num1Index]
        num2 = pickNumbers[num2Index]
        if num1 < num2:
            pickNumbers[num1Index], pickNumbers[num2Index] = pickNumbers[num2Index], pickNumbers[num1Index]
            pickList[num1Index], pickList[num2Index] = pickList[num2Index], pickList[num1Index]
            winNumbers[num1Index], winNumbers[num2Index] = winNumbers[num2Index], winNumbers[num1Index]
        num2Index += 1
    num1Index += 1
        
for pick in pickList:
    print(pick)
    print(pickNumbers[pickList.index(pick)])
    print(winNumbers[pickList.index(pick)])
print(len(pickList))
```
