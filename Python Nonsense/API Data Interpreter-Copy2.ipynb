```python
playerIds = []
numGames = []
winNumbers = []
killNumbers = []
deaths = []
assists = []
rank = []
damageNumbers = []
iterator = 0

for Id in df['Player Id']:
    if Id != 0:
        if Id in playerIds:
            numGames[playerIds.index(Id)] += 1
            if df['Win Status'].values[iterator] == '"Winner"':
                winNumbers[playerIds.index(Id)] += 1
            killNumbers[playerIds.index(Id)] += df['Player Kills'].values[iterator]
            deaths[playerIds.index(Id)] += df['Deaths'].values[iterator]
            assists[playerIds.index(Id)] += df['Assists'].values[iterator]
            damageNumbers[playerIds.index(Id)] += df['Player Damage'].values[iterator]
        else:
            playerIds.append(Id)
            rank.append(df['Conquest Rank'].values[iterator])
            numGames.append(1)
            if df['Win Status'].values[iterator] == '"Winner"':
                winNumbers.append(1)
            else:
                winNumbers.append(0)
            killNumbers.append(df['Player Kills'].values[iterator])
            deaths.append(df['Deaths'].values[iterator])
            assists.append(df['Assists'].values[iterator])
            damageNumbers.append(df['Player Damage'].values[iterator])
        print(df['God name'].values[iterator])
        iterator += 1

iterator = 0
while iterator < len(playerIds):
    if numGames[iterator] < 5:
        playerIds.pop(iterator)
        numGames.pop(iterator)
        winNumbers.pop(iterator)
        killNumbers.pop(iterator)
        deaths.pop(iterator)
        assists.pop(iterator)
        rank.pop(iterator)
        damageNumbers.pop(iterator)
    else:
        iterator += 1
        
        
num1Index = 0       
while num1Index < len(playerIds) - 1:
    num2Index = num1Index + 1
    while num2Index < len(playerIds):
        num1 = winNumbers[num1Index] / numGames[num1Index]
        num2 = winNumbers[num2Index] / numGames[num2Index]
        if num1 < num2:
            playerIds[num1Index], playerIds[num2Index] = playerIds[num2Index], playerIds[num1Index]
            numGames[num1Index], numGames[num2Index] = numGames[num2Index], numGames[num1Index]
            winNumbers[num1Index], winNumbers[num2Index] = winNumbers[num2Index], winNumbers[num1Index]
            killNumbers[num1Index], killNumbers[num2Index] = killNumbers[num2Index], killNumbers[num1Index]
            deaths[num1Index], deaths[num2Index] = deaths[num2Index], deaths[num1Index]
            assists[num1Index], assists[num2Index] = assists[num2Index], assists[num1Index]
            rank[num1Index], rank[num2Index] = rank[num2Index], rank[num1Index]
            damageNumbers[num1Index], damageNumbers[num2Index] = damageNumbers[num2Index], damageNumbers[num1Index]
        num2Index += 1
    num1Index += 1
```
